```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('test.png')
```

```python
y = x * x
plt.plot(x,y,'r*--')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Graph in 2D")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(x,y,'g--')

plt.subplot(2,2,3)
plt.plot(x,y,'b*-')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title('Matplotlib Demo')
plt.xlabel('X axis caption')
plt.ylabel('Y axis caption')
plt.plot(x,y,'r*-')
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title('Sine wave form')
plt.plot(x,y,'g')
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.cos(x)
plt.title('Cosine wave form')
plt.plot(x,y,'b')
```

```python
plt.subplot(2,1,1)
plt.title('Cosine')
plt.plot(x,y,'b')

y = np.sin(x)
plt.subplot(2,1,2)
plt.plot(x,y,'y')
plt.title('Sine')
```

```python
x = np.arange(0,4)
y = np.arange(10,14)
plt.bar(x, y, color = 'g')

x2 = np.arange(5,9)
y2 = np.arange(15,19)
plt.bar(x2, y2, color = 'b')
```
