### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

#### Scatterplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px      # importing this only for a dataset
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# scatterplot (axis level function)
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
# relplot (figure level function)
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter')
```

```python
# amazing features of seaborn
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'sex')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter',hue = 'sex', style = 'time')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter',hue = 'sex', style = 'time', size = 'size')
```

#### Lineplot

```python
df = px.data.gapminder()  # importing dataset from plotly
df = df[ df['country'] == 'Bangladesh' ]
df
```

```python
# lineplot (axis level function)
sns.lineplot(data = df, x = 'year', y = 'lifeExp')
```

```python
# relplot (figure level function)
sns.relplot(kind = 'line', data = df, x = 'year', y = 'lifeExp')
```

```python
df = px.data.gapminder()  # importing dataset from plotly
df = df[ df['country'].isin(['Bangladesh', 'India', 'Australia', 'Canada']) ]
df
```

```python
# hue, style
sns.relplot(kind = 'line', data = df, x = 'year', y = 'lifeExp', hue = 'country', style = 'continent')
```

```python
# facetplot can be only used in figure level function
sns.relplot(kind = 'line', data = df, x = 'year', y = 'lifeExp', hue = 'country', style = 'continent', col = 'country', row = 'continent')
```

```python
# col_wrap
sns.relplot(kind = 'line', data = df, x = 'year', y = 'lifeExp', hue = 'country', style = 'continent', col = 'country', col_wrap = 2)
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

#### Histplot

```python
tips
```

```python
# histplot (axis level function)
sns.histplot(data = tips, x = 'total_bill')
```

```python
# displot (figure level function)
sns.displot(kind = 'hist', data = tips, x = 'total_bill')
```

```python
# bins parameter
sns.displot(kind = 'hist', data = tips, x = 'total_bill', bins = 5)
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables

# but it is basically a count plot
sns.histplot(data = tips, x = 'day')
```

```python
# hue
sns.displot(kind = 'hist', data = tips, x = 'total_bill', bins = 5, hue = 'sex')
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(kind = 'hist', data = tips, x = 'total_bill', bins = 5, col = 'sex')
```

```python
# element
sns.displot(kind = 'hist', data = tips, x = 'total_bill', bins = 5, hue = 'sex', element = 'step')
```

#### KDEplot

```python
tips
```

```python
# kdeplot (axis level function)
sns.kdeplot(data = tips, x = 'total_bill')
```

```python
# displot (figure level)
sns.displot(kind = 'kde', data = tips, x = 'total_bill')
```

```python
sns.displot(kind = 'kde', data = tips, x = 'total_bill', hue = 'sex')
```

```python
sns.displot(kind = 'kde', data = tips, x = 'total_bill', hue = 'sex', fill = True)
```

```python
sns.displot(kind = 'kde', data = tips, x = 'total_bill', col = 'sex')
```

#### Rugplot

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.displot(kind = 'kde', data = tips, x = 'total_bill')
sns.rugplot(data = tips, x = 'total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(data=tips, x='total_bill', y='tip')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
df = px.data.gapminder()
df
```

```python
# heatmap (axis level function)
temp_df = df.pivot(index = 'country', columns = 'year', values = 'lifeExp')
sns.heatmap(temp_df)
```

```python
temp_df = df[ df['continent'] == 'Asia' ].pivot(index = 'country', columns = 'year', values = 'lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(temp_df, cmap = 'summer', annot = True, linewidth = 1)
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,:4])
```
