```python
class Item:
    items = []
    payrate = 0.8 # it means 20 percent off
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
        
        Item.items.append(self)
        
    def print_details(self):
        return f"Name : {self.name}\nPrice : {self.price}\nQuantity : {self.quantity}"
    
    def calculate_total_price(self):
        return self.price * self.quantity
    
    def apply_discount(self):
        self.price = self.price * self.payrate
    
item1 = Item("Soap",160,2)
print(item1.print_details())
item1.apply_discount()
print(item1.print_details())

item2 = Item("Shampoo",60,3)
print(item2.print_details())
item2.payrate = 0.7  # it means 30 percent off
item2.apply_discount()
print(item2.print_details())

for item in Item.items:
    print(item.print_details())
    
    
```

```python
class Phone(Item):
    def __init__(self,name, price, quantity, broken_phones):
        super().__init__(name, price, quantity)
        self.quantity = quantity - broken_phones

phone_1 = Phone("Xiaomi", 15000, 3, 1)
print(phone_1.calculate_total_price())
print(phone_1.print_details())
```

```python
class Employee:
    raise_amount = 1.04
    no_of_emps = 0
    def __init__(self, first, last, salary):
        self.first = first
        self.last = last
        self.mail = self.first + self.last.upper() + "@gmail.com"
        self.salary = salary
        Employee.no_of_emps += 1
        
    def details(self):
        return f"First Name : {self.first}\nLast Name : {self.last}\nEmail : {self.mail}\nSalary : {self.salary}"
    
    @classmethod
    def set_raise_amt(cls, amount):
        cls.raise_amount = amount
    
    @classmethod
    def from_string(cls, string):
        first, last, salary = string.split("-")
        return cls(first, last, salary)
    
    def apply_raise(self):
        self.salary *= self.raise_amount
    @staticmethod
    def is_off_day(day):
        if day.weekday() == 2 or day.weekday() == 6:   # weekday 0 = monday, weekday 6 = sunday
            return True
        return False
    
emp1 = Employee("Sadikur", "Siam", 150000)
emp2 = Employee("Roman", "Nihal", 160000)
print(emp1.details())
print(emp2.details())

#print(emp1.raise_amount)
#print(emp2.raise_amount)
#Employee.set_raise_amt(1.10)
#print(emp1.raise_amount)
#print(emp2.raise_amount)

#emp1.raise_amount = 1.20
#emp1.apply_raise()
#emp2.apply_raise()
#print(emp1.details())
#print(emp2.details())
#print(Employee.no_of_emps)
#print(emp1.no_of_emps)
#print(emp2.no_of_emps)


#print(emp1.raise_amount)
#print(emp2.raise_amount)
emp_string = "Abu-sayem-170000"
emp3 = Employee.from_string(emp_string)
print(emp3.details())
#print(Employee.no_of_emps)


from datetime import date
today = date.today()
#print(today)
print(Employee.is_off_day(today))
```

```python
class Developer(Employee):
    def __init__(self, first, last, salary, prog_lang):
        super().__init__(first, last, salary)
        self.prog_lang = prog_lang
    def __repr__(self):
        return f"Developer('{self.first}', '{self.last}', {self.salary}, '{self.prog_lang}')"
    
    def __str__(self):
        return f"{self.first} {self.last}- {self.mail}"
dev_1 = Developer("Omor", "Faruk", 200000, 'Python')
dev_2 = Developer("Nahid", "Ahmed", 210000, 'Java')
print(dev_1.details())
print(dev_2.details())

class Manager(Employee):
    def __init__(self, first, last, salary, employees):
        super().__init__(first, last, salary)
        if employees == None:
            self.employees = []
        else:
            self.employees = employees
            
    def add_emp(self, emp):
        if emp not in self.employees:
            self.employees.append(emp)
            
    def remove_emp(self,emp):
        if emp in self.employees:
            self.employees.remove(emp)
            
    def employees_under_manager(self):
        for emp in self.employees:
             print(f"-->: {emp.first} {emp.last}")
            
mng1 = Manager("Solaiman","Mia",500000,[])
mng2 = Manager("Aminur","Rahman",600000,[dev_1,dev_2])

print(mng1.details())
mng1.employees_under_manager()
print(mng2.details())
mng2.employees_under_manager()
mng2.remove_emp(dev_2)
mng2.employees_under_manager()
mng2.add_emp(dev_2)
mng2.employees_under_manager()

print(dev_1)
print(dev_2.__repr__())
```
