```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips') 
# The tips dataset is one of the example datasets provided by the Seaborn library itself. 
# It's a built-in dataset that contains information about tips given in a restaurant. This dataset is often used for
# demonstrating various visualization techniques and statistical analyses in Seaborn.
#df.head()
df.iloc[:,:]
```

### Correlation with heatmap

```python
corr = df.corr(numeric_only=True)
print(corr)
```

```python
sns.heatmap(corr)
```

```python
sns.heatmap(corr, cmap = 'RdBu', vmax = 1, vmin = -1)
```

### Joint Plot

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'reg')
```

### Pair Plot

```python
sns.pairplot(df)
```

### Line Plot

```python
var = [2, 6, 6]
var_1 = [1, 3, 5]

df = pd.DataFrame({"var":var, "var_1":var_1})

sns.lineplot(x = 'var_1', y = 'var', data = df)
```

```python
df = sns.load_dataset('penguins')
df = df.head(10)
print(df)
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'flipper_length_mm', data = df)
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'flipper_length_mm', data = df, hue = 'sex')
```

### Dist plot

```python
df = sns.load_dataset('penguins')
sns.displot(df['flipper_length_mm'])
```

```python
sns.displot(df['flipper_length_mm'],bins=20,kde=True,rug=True,color='r')
```
